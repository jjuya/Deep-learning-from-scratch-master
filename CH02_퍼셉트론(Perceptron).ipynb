## 퍼셉트론 간단 구현

### AND 게이트

```python
def AND(x1, x2): 
    w1, w2, theta = 0.5, 0.5, 0.7 
    tmp = w1*x1 + w2*x2 
    if tmp <= theta: return 0 
    else: return 1
```

### 바이어스(bias, 편향) 도입



- 입력값이 출력에 어떠한 영향을 미치는지 알기 위한 값
- 바이어스는 출력에 대한 default 값
- x0 = 1 이면 선형 회귀와 똑같다

```python
import numpy as np

def AND(x1, x2): 
    x = np.array([x1, x2]) 
    w = np.array([0.5, 0.5])
    b = -0.7 
    
    tmp = np.sum(w*x) + b 
    
    if tmp <= 0: return 0 
    else: return 1
    
```

```python
for xs in [(0, 0), (1, 0), (0, 1), (1, 1)]:
    y = AND(xs[0], xs[1]) 
    print(str(xs) + " -> " + str(y))
```

```python
import numpy as np 

def NAND(x1, x2): 
    x = np.array([x1, x2]) 
    w = np.array([-0.5, -0.5]) 
    b = 0.7 
    
    tmp = np.sum(w*x) + b 
    
    if tmp <= 0: return 0 
    else: return 1
```

```python
for xs in [(0, 0), (1, 0), (0, 1), (1, 1)]:
    y = AND(xs[0], xs[1]) 
    print(str(xs) + " -> " + str(y))
```

```python
import numpy as np 

def OR(x1, x2): 
    x = np.array([x1, x2]) 
    w = np.array([1.0, 1.0]) 
    b = -0.5 
    
    tmp = np.sum(w*x) + b 
    
    if tmp <= 0: return 0 
    else: return 1
```

```python
for xs in [(0, 0), (1, 0), (0, 1), (1, 1)]:
    y = AND(xs[0], xs[1]) 
    print(str(xs) + " -> " + str(y))
```

### 다층 퍼셉트론(MLP)


```python
def XOR(x1, x2): 
    s1 = NAND(x1, x2) 
    s2 = OR(x1, x2) 
    y = AND(s1, s2) 
    
    return y
```

```python
for xs in [(0, 0), (1, 0), (0, 1), (1, 1)]: 
    y = XOR(xs[0], xs[1]) 
    print(str(xs) + " -> " + str(y))
```
